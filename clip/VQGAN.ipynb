## Generating Images with CLIP: VQGAN + CLIP

Because CLIP can provide a metric of how well the image suits a textual description, we can construct a neural architecture that will consist of a **generator** that learns how to produce images, and **CLIP descriminator**, which determines how close the image is to the desired textual description. By doing so, we will make AI produce a picture from text!

This sounds simple, but the details a quite complex. We will use existing library called [pixray](https://github.com/pixray/pixray), and borrow the [example](https://github.com/pixray/pixray_notebooks) from there.

First, we will do some installations (and you need to restart kernel after this is done):

```python
import sys
sys.path.append("pixray")
import pixray
```

```python
prompts = "beautiful girl with a robotic cat"

pixray.reset_settings()
pixray.add_settings(prompts=prompts)
pixray.add_settings(quality="better")
pixray.add_settings(drawer="vqgan")
pixray.add_settings(custom_loss="saturation:2,smoothness:1.5")
pixray.add_settings(smoothness_type="log")
pixray.add_settings(display_clear=True)
settings = pixray.apply_settings()
pixray.do_init(settings)
pixray.do_run(settings)
```
